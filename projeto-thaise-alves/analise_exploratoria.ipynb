```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('salas-de-exibicao-e-complexos.csv', sep=';')
```

```python
df
```

```python
df_filtrado = df[(df['SITUACAO_SALA'] == 'EM FUNCIONAMENTO') & (df['SITUACAO_EXIBIDOR'] == 'REGULAR')]
df_filtrado
```

```python
df_colunas_selecionadas = df_filtrado[['NOME_SALA', 'SITUACAO_SALA', 'DATA_SITUACAO_SALA', 'ASSENTOS_SALA', 'ASSENTOS_CADEIRANTES', 'ASSENTOS_MOBILIDADE_REDUZIDA', 'ASSENTOS_OBESIDADE', 'ACESSO_ASSENTOS_COM_RAMPA', 'ACESSO_SALA_COM_RAMPA', 'BANHEIROS_ACESSIVEIS', 'MUNICIPIO_COMPLEXO', 'CEP_COMPLEXO', 'UF_COMPLEXO', 'SITUACAO_EXIBIDOR']]
df_colunas_selecionadas
```

```python
nulos_por_coluna = df_colunas_selecionadas.isnull().sum()
nulos_por_coluna
```

```python
df_sem_nulos = df_colunas_selecionadas.dropna()
df_sem_nulos
```

```python
nulos_por_coluna = df_sem_nulos.isnull().sum()
nulos_por_coluna
```

```python
df_colunas_selecionadas.describe(include='all').T
```

```python
df_colunas_selecionadas.to_csv('base_final_salas_e_complexos_de_cinema.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8, 6))

sns.countplot(data = df_sem_nulos, x = 'ACESSO_SALA_COM_RAMPA')

plt.title('O acesso para as salas de cinema possui rampa')
plt.ylabel('Contagem')

plt.show()
```

```python
plt.figure(figsize=(8, 6))

sns.countplot(data = df_sem_nulos, x = 'BANHEIROS_ACESSIVEIS')

plt.title('Os banheiros nos cinemas possuem acessibilidade')
plt.ylabel('Contagem')

plt.show()
```

```python
plt.figure(figsize=(8, 6))

sns.countplot(data = df_sem_nulos, x = 'ACESSO_ASSENTOS_COM_RAMPA')

plt.title('O acesso aos assentos possuem rampa')
plt.ylabel('Contagem')

plt.show()
```
