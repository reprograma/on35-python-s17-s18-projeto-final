```python
df = pd.read_csv('microdados_ed_basica_2023.csv', delimiter=';', encoding='latin1')
```

```python
df_ba = df[df['SG_UF'] == 'BA'] 
```

```python
colunas_manter = ['NU_ANO_CENSO','NO_MUNICIPIO','NO_ENTIDADE','TP_DEPENDENCIA',
'TP_CATEGORIA_ESCOLA_PRIVADA','TP_LOCALIZACAO','TP_SITUACAO_FUNCIONAMENTO',
'IN_INTERNET','IN_INTERNET_ALUNOS','IN_DESKTOP_ALUNO','QT_DESKTOP_ALUNO', 'IN_COMP_PORTATIL_ALUNO', 'QT_COMP_PORTATIL_ALUNO', 'IN_TABLET_ALUNO', 'QT_TABLET_ALUNO',
'IN_FORMA_CONT_ES_PREST_SERV','IN_FORMA_CONT_ES_COOP_TEC_FIN','IN_FORMA_CONT_ES_CONSORCIO_PUB',
'TP_AEE','IN_MEDIACAO_PRESENCIAL','IN_MEDIACAO_SEMIPRESENCIAL',
'IN_MEDIACAO_EAD','IN_DIURNO','IN_NOTURNO','IN_MED','IN_PROF','IN_PROF_TEC','QT_MAT_BAS_FEM','QT_MAT_BAS_MASC',
'QT_MAT_BAS_ND','QT_MAT_BAS_BRANCA','QT_MAT_BAS_PRETA','QT_MAT_BAS_PARDA','QT_MAT_BAS_AMARELA','QT_MAT_BAS_INDIGENA']
```

```python
df_ba_filtrado = df_ba[colunas_manter]
```

```python
#seleção de colunas do banco de dados
df_ba_filtrado.columns = ['ano censo', 'nome do municipio', 'nome da escola', 'dependência administrativa',
'categoria escola privada', 'localização', 'situação de funcionamento', 'internet', 'internet para alunos','computador', 'qt computador','computador portatil', 'qt computador portatil', 'tablet', 'qt tablet',
'contrato de prestação de serviço', 'termo de cooperação técnica e financeira', 'tipo de atendimento escolarização',
'Atendimento Educacional Especializado (AEE)', 'Mediação Presencial', 'Mediação Semipresencial', 'Mediação EAD', 'Turno Diurno', 'Turno Noturno', 'Ensino Médio', 'Educação Profissional', 'Educação Profissional Técnica', 'Feminino', 'Masculino', 'Raça não Declarada', 'Branca', 'Preta', 'Parda', 'Amarela', 'Indígena']
```

```python
df_ba_filtrado
```

```python
df_ba_filtrado['dependência administrativa'] = df_ba_filtrado['dependência administrativa'].replace({
    1:'federal', 
    2:'estadual', 
    3:'municipal', 
    4:'privada'})
```

```python
df_ba_filtrado['categoria escola privada'] = df_ba_filtrado['categoria escola privada'].replace({
    1:'particular', 
    2:'comunitaria', 
    3:'confessional', 
    4:'filantropica'})
```

```python
df_ba_filtrado['localização'] = df_ba_filtrado['localização'].replace({
    1:'urbana', 
    2:'rural'})
```

```python
df_ba_filtrado['situação de funcionamento'] = df_ba_filtrado['situação de funcionamento'].replace({
    1:'ativa',
    2:'paralisada',
    3:'extinta no ano do censo',
    4:'extinta anos antes do censo'
})
```

```python
df_ba_filtrado['internet'] = df_ba_filtrado['internet'].replace({
    0:'Não',
    1:'Sim'
})
```

```python
df_ba_filtrado['internet para alunos'] = df_ba_filtrado['internet para alunos'].replace({
    0:'Não',
    1:'Sim'
})
```

```python
df_ba_filtrado['computador'] = df_ba_filtrado['computador'].replace({
    0:'Não',
    1:'Sim'
})
df_ba_filtrado['computador portatil'] = df_ba_filtrado['computador portatil'].replace({
    0:'Não',
    1:'Sim'
})
df_ba_filtrado['tablet'] = df_ba_filtrado['tablet'].replace({
    0:'Não',
    1:'Sim'
})
```

```python
df_ba_filtrado[['computador','computador portatil', 'tablet','internet', 'internet para alunos']] = df_ba_filtrado[['computador', 'computador portatil', 'tablet', 'internet', 'internet para alunos']].fillna('não informado')
```

```python
df_ba_filtrado
```

```python
#Filtro para escolas com ensino médio
df_ba_filtrado = df_ba_filtrado.drop(df_ba_filtrado[df_ba_filtrado['Ensino Médio'] != 1].index)
```
