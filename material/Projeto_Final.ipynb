```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('tabela para formatar (1).csv', delimiter=';', encoding='latin1')
```

```python
#visualizando dataframe
df.head(10000)
```

```python
#visualizando colunas do dataframe
print(df.columns)
```

```python
#Deletando colunas que não irei usar
df = df.drop(['class_level_1', 'class_level_0','feature_id','class_level_3', 'class_level_4', 'class'], axis=1)
```

```python
#Verificando se as colunas foram deletadas
print(df.columns)
```

```python
df.head(10000)
```

```python
df_melted = df.melt(id_vars=['ï»¿country', 'state', 'municipality', 'municipality - state', 'geocode', 'class_level_2'],
                    var_name='Ano', value_name='Valor')
df_melted['Valor'] = df_melted['Valor'].astype(str).str.replace(',', '.').astype(float)

# Passo 2: Pivotear para que cada class_level_2 vire uma coluna
df_pivot = df_melted.pivot_table(index=['ï»¿country', 'state', 'municipality', 'municipality - state', 'geocode', 'Ano'],
                                 columns='class_level_2',
                                 values='Valor',
                                 fill_value=0)

# Ajustando para deixar em formato de DataFrame
df_pivot.reset_index(inplace=True)

# Exibindo o resultado final
print("\nTabela Transformada:")
print(df_pivot)
```

```python
df_pivot.columns
```

```python
df_pivot.head(100000)
```

```python
df_pivot.isnull().sum()
```

```python
df_pivot['Estado'] = df_pivot['Estado'].replace(['RondÃ´nia'], 'Rondônia')
```

```python
df_pivot.head(100000)
```

```python
pd.options.display.float_format = '{:,.2f}'.format
```

```python
df_pivot.rename(columns={
    'ï»¿country': 'País',
    'state': 'Estado',
    'municipality': 'Município',
    'municipality - state': 'Município - Estado',
    'geocode': 'Geocódigo',
    'feature_id': 'ID Da Característica',
    'Ano': 'Ano',
    '1.1. Forest Formation': 'Formação Florestal',
    '1.2. Savanna Formation': 'Formação De Savana',
    '1.4 Floodable Forest': 'Floresta Alagável',
    '2.1. Wetland': 'Área Pantanosa',
    '2.2. Grassland': 'Formação Campestre',
    '2.4. Rocky Outcrop': 'Afloramento Rochoso',
    '3.1. Pasture': 'Pastagem',
    '3.2. Agriculture': 'Agricultura',
    '3.3. Forest Plantation': 'Silvicultura',
    '3.4. Mosaic of Uses': 'Mosaico De Usos',
    '4.2. Urban Area': 'Área Urbana',
    '4.3. Mining': 'Mineração',
    '4.4. Other non Vegetated Areas': 'Outras Áreas Não Vegetadas',
    '5.1. River, Lake and Ocean': 'Rio, Lago E Oceano',
    '5.2. Aquaculture': 'Aquicultura',
    '6. Not Observed': 'Não Observado'
}, inplace=True)
```

```python
df_pivot.columns
```

```python
# Criando gráfico
plt.figure(figsize=(10, 5))
plt.bar(df_pivot['Estado'], df_pivot['Formação Florestal'], color='green')
plt.title('Área de Formação Florestal por Estado')
plt.xlabel('Estados')
plt.ylabel('Área de Formação Florestal (hectares)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(df_pivot['Estado'], df_pivot['Agricultura'], color='brown')  # Gráfico de barras

plt.title('Área de Agricultura por Estado')
plt.xlabel('Estados')
plt.ylabel('Área de Agricultura (hectares)')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
