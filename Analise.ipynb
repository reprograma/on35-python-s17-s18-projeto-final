# Mapeamento

```python
raca_cor_map = {
    1: 'Branca',
    2: 'Preta',
    3: 'Parda',
    4: 'Amarela',
    5: 'Indígena',
    6: 'Não informada',
    9: 'Não Identificado'
}

df['raca_cor'] = df['raca_cor'].replace(raca_cor_map)

```

```python
print(raca_cor_map)
```

# Graficos

```python
df['raca_cor'] = df['raca_cor'].replace(raca_cor_map)

# Gráfico da distribuição de raça
plt.figure(figsize=(10, 6))
df['raca_cor'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribuição de Raça')
plt.xlabel('Raça')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

```python
# Gráfico da distribuição de sexo
plt.figure(figsize=(8, 5))
contagem = df['sexo_descricao'].value_counts()
contagem.plot(kind='bar', color=['#3cacae', '#c26dbc'])  
plt.title('Distribuição de Sexo')
plt.xlabel('Sexo')
plt.ylabel('Quantidade')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.pie(contagem, labels=contagem.index, autopct='%1.1f%%', startangle=90, colors=['#3cacae', '#c26dbc'])
plt.title('Proporção de Sexo')
plt.axis('equal') 
plt.show()
```

```python

media_salarial = df.groupby(['raca_cor', 'sexo_descricao'])['salario_mensal'].mean().unstack()


plt.figure(figsize=(10, 6))
media_salarial.plot(kind='bar', color=['#3cacae', '#c26dbc'])
plt.title('Média Salarial por Raça e Sexo')
plt.xlabel('Raça')
plt.ylabel('Média Salarial')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Sexo')
plt.show()
```

```python
media_salarial_uf = df.groupby(['sigla_uf', 'sexo_descricao'])['salario_mensal'].mean().unstack()



plt.figure(figsize=(15, 15))
media_salarial_uf.plot(kind='bar', color=['#3cacae', '#c26dbc'], edgecolor='black')
plt.title('Média Salarial por Estado e Sexo')
plt.xlabel('Estado')
plt.ylabel('Média Salarial')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Sexo')
plt.tight_layout()  
plt.show()
```
