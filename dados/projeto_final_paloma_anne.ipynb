```python
# Importando os dados das planilhas

# Primeiro, instale a biblioteca pandas se ainda não estiver instalada
!pip install pandas

# Importando os dados
import pandas as pd
import csv

# Carregando os dois arquivos CSV
# Caso esteja rodando localmente, altere o caminho dos arquivos, no Colab, você pode fazer upload dos arquivos e colocar o nome do arquivo abaixo:

df1 = pd.read_csv('SEEG.csv')


# Exibindo as primeiras linhas dos DataFrames para verificar
print(df1.head())

```

```python

df1.info()
```

```python
# Selecionar as colunas específicas

df_1 = df1[['Categoria','2019', '2020', '2021', '2022']]
print(df_1)
```

```python
# Altera os nomes das colunas

df_1.columns = ['Municípios', '2019', '2020', '2021', '2022']
print(df_1)
```

```python
# Remover informações de estado (como "(BA)") de 'Municípios'
df_1['Municípios'] = df_1['Municípios'].str.replace(r' \(.*\)', '', regex=True)

print(df_1)
```

```python
# Converta todas as colunas de anos para inteiro

df_1[['2019', '2020', '2021', '2022']] = df_1[['2019', '2020', '2021', '2022']].astype(int)

print(df_1)
```

```python
# Contagem de valores nulos em cada coluna

nulos_por_coluna1 = df_1.isnull().sum()
print(nulos_por_coluna1)
```

```python
# Remove linhas duplicadas

df_sem_duplicatas1 = df_1.drop_duplicates()
print(df_sem_duplicatas1)
```

```python
 # fazer se quiser conferir os dados

# Calcular o total da coluna

valor_total = df_1['2019'].sum()
print("O valor total da coluna é:", valor_total)

valor_total = df_1['2020'].sum()
print("O valor total da coluna é:", valor_total)

valor_total = df_1['2021'].sum()
print("O valor total da coluna é:", valor_total)

valor_total = df_1['2022'].sum()
print("O valor total da coluna é:", valor_total)


df_1.info()
```

```python
# Transformar os dados para um formato longo (melt), os 4 anos para uma coluna

df_sem_duplicatas1 = pd.melt(df_1, id_vars=['Municípios'], value_vars=['2019', '2020', '2021', '2022'], var_name='Ano', value_name='CO2_emitido')

print(df_sem_duplicatas1)
```

```python
# fazer somente para conferir os dados


df1_MUN = df_sem_duplicatas1.groupby('Municípios')['CO2_emitido'].sum().reset_index()

# Mostrar o resultado
print(df1_MUN.head())

df1_ANO = df_sem_duplicatas1.groupby('Ano')['CO2_emitido'].sum().reset_index()

# Mostrar o resultado
print(df1_ANO.head())
```

```python
# conhecendo o dado

# Conta os códigos únicos de municípios
num_municipios = df_sem_duplicatas1['Municípios'].nunique()

print(f"Existem {num_municipios} municípios únicos no arquivo CSV.")
```

```python
# conhecendo o dado

# Obtém os códigos únicos de municípios
municipios_unicos = df_sem_duplicatas1['Municípios'].unique()

# Exibe os códigos dos municípios únicos
print(municipios_unicos)
```

```python
# Verificar se 'OUTROS' está presente nos dados
outros_presentes = df_sem_duplicatas1[df_sem_duplicatas1['Municípios'].str.contains('OUTROS', case=False, na=False)]
print("Linhas com 'OUTROS' ")
print(outros_presentes)
```

```python
# Salvando o arquivo final em um novo CSV

df_sem_duplicatas1.to_csv('co2.csv', index=False, encoding='utf-8')

print(df_sem_duplicatas1)

```

```python
# Importar bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar os dados integrados
df_co2 = pd.read_csv('co2.csv')

# Exibir as primeiras linhas do DataFrame para verificar se os dados foram carregados corretamente
print(df_co2.head())
```

```python
# conferindo os dados

df_MUN = df_co2.groupby('Municípios')['CO2_emitido'].sum().reset_index()

# Mostrar o resultado
print(df_MUN.head())

df_ANO = df_co2.groupby('Ano')['CO2_emitido'].sum().reset_index()

# Mostrar o resultado
print(df_ANO.head())


# Somar os valores da coluna AREAHA
total_co2 = df_co2['CO2_emitido'].sum()

print(f'Total : {total_co2:.2f}')
```

```python
# Definir o estilo para os gráficos
sns.set(style="whitegrid")
```

```python
# Agrupar os dados por ano e somar as emissões
grouped = df_co2.groupby('Ano')['CO2_emitido'].sum().reset_index()

# Criar o gráfico de barras para CO₂ emitido
plt.figure(figsize=(10, 6))

# Definir uma paleta de cores distinta
colors = sns.color_palette("husl", len(grouped['Ano'].unique()))

# Criar o gráfico de barras, utilizando as cores definidas
sns.barplot(x='Ano', y='CO2_emitido', data=grouped, palette=colors, errorbar=None)

plt.title('CO₂ emitido por Ano na Bahia', fontsize=16)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Emissões de CO₂ (ton)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()  # Ajusta o layout para evitar sobreposição
plt.show()
```

```python

# Filtrar para remover os municípios classificados como 'outros'
df_co2 = df_co2[df_co2['Municípios'] != 'Outros']

# Agrupar por ano e município, somando as emissões
grouped = df_co2.groupby(['Ano', 'Municípios'])['CO2_emitido'].sum().reset_index()

# Filtrar para remover municípios com emissão zero
grouped = grouped[grouped['CO2_emitido'] > 0]

# Criar gráficos para cada ano
years = grouped['Ano'].unique()  # Obter anos únicos

plt.figure(figsize=(12, 10))

# Definir uma paleta de cores distinta
colors = sns.color_palette("husl", len(years))

for i, year in enumerate(years):
    # Filtrar os 10 maiores municípios para cada ano
    top_10_municipios = grouped[grouped['Ano'] == year].nlargest(10, 'CO2_emitido')

    # Criar um subgráfico
    plt.subplot(len(years), 1, i + 1)

    # Criar o gráfico de barras
    sns.barplot(x='CO2_emitido', y='Municípios', data=top_10_municipios,
                palette=[colors[i]], dodge=False)

    plt.title(f'Top 10 Municípios com Mais Emissões de CO₂ em {year}', fontsize=14)
    plt.xlabel('Emissões de CO₂ (ton)', fontsize=12)
    plt.ylabel('Município', fontsize=12)
    plt.xticks(rotation=45)

    # Adicionar a legenda específica para cada gráfico
    handles = [plt.Line2D([0], [0], color=colors[i], lw=4)]
    plt.legend(handles, [year], title='Ano', frameon=False)

plt.tight_layout()
plt.show()

```

```python

# Agrupar por ano e município, somando as áreas desmatadas
grouped = df_desmatamento.groupby(['ANODETEC', 'MUNICIPIO'])['AREAHA'].sum().reset_index()

# Filtrar para remover municípios com área desmatada zero
grouped = grouped[grouped['AREAHA'] > 0]

# Obter anos únicos
years = grouped['ANODETEC'].unique()

plt.figure(figsize=(12, 10))

# Definir uma paleta de cores distinta
colors = sns.color_palette("husl", len(years))

for i, year in enumerate(years):
    # Filtrar os 10 maiores municípios para cada ano
    top_10_municipios = grouped[grouped['ANODETEC'] == year].nlargest(10, 'AREAHA')

    # Criar um subgráfico
    plt.subplot(len(years), 1, i + 1)

    # Criar o gráfico de barras
    sns.barplot(x='AREAHA', y='MUNICIPIO', data=top_10_municipios,
                palette=[colors[i]], dodge=False)

    plt.title(f'Top 10 Municípios com Mais Desmatamento em {year}', fontsize=14)
    plt.xlabel('Área Desmatada (ha)', fontsize=12)
    plt.ylabel('Município', fontsize=12)
    plt.xticks(rotation=45)

    # Adicionar a legenda específica para cada gráfico
    handles = [plt.Line2D([0], [0], color=colors[i], lw=4)]
    plt.legend(handles, [year], title='Ano', frameon=False)

plt.tight_layout()
plt.show()
```
