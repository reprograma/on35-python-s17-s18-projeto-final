### Análise de Dados com Gráficos

### Qual é o número total de incidentes de violência contra mulheres registrados em cada bairro de Vila Velha durante os anos de 2022 e 2023?

```python
# Contagem do número de incidentes de violência contra mulheres por bairro e armazena os resultados em 'bairro_counts' 
bairro_counts = df_violencia_mulheres['BAIRRO'].value_counts()
```

```python
import matplotlib.pyplot as plt

top_n = 20
bairro_counts_top = bairro_counts.nlargest(top_n)

plt.figure(figsize=(12, 6))

cor_barras = '#CDA2D7'  
cor_borda = '#6D4C71' 
cor_texto = '#333333'  
cor_fundo = '#EAEAEA'

bars = plt.bar(bairro_counts_top.index, bairro_counts_top.values, color=cor_barras, edgecolor=cor_borda)

plt.xticks(rotation=90, color=cor_texto)
plt.title('20 Bairros Mais Incidentes de Violência contra Mulheres em Vila Velha (2022-2023)', fontsize=14, color=cor_texto)
plt.xlabel('Bairro', fontsize=12, color=cor_texto)
plt.ylabel('Número de Incidentes', fontsize=12, color=cor_texto)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom', color=cor_texto)

plt.gca().set_facecolor(cor_fundo)

plt.tight_layout()

plt.show()
```

### Qual é a quantidade de incidentes de violência contra mulheres em Vila Velha, categorizados por tipo de local em 2022 e 2023?

```python
# Contagem do número de incidentes de violência contra mulheres por tipo de local, transformando os resultados em um DataFrame com colunas renomeadas para 'TIPO DE LOCAL' e 'contagem'
local_counts = df_violencia_mulheres['TIPO DE LOCAL'].value_counts().reset_index()
local_counts.columns = ['TIPO DE LOCAL', 'contagem']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 8))

cor_principal = '#CDA2D7' 
cor_destaque = '#6D4C71'  
cor_fundo = '#EAEAEA'  
cor_texto = '#333333'  

palette = [cor_principal, cor_destaque]

sns.barplot(data=local_counts, x='TIPO DE LOCAL', y='contagem', palette=palette)

plt.xticks(rotation=45, color=cor_texto)  
plt.title('Incidentes de Violência contra Mulheres por Tipo de Local em Vila Velha (2022-2023)', fontsize=16, color=cor_texto)
plt.xlabel('Tipo de Local', fontsize=14, color=cor_texto)
plt.ylabel('Número de Incidentes', fontsize=14, color=cor_texto)

plt.gca().set_facecolor(cor_fundo)

plt.tight_layout()  

plt.show()
```

### Como estão distribuídos os incidentes de violência contra mulheres em Vila Velha, classificados por código e tipo de incidente, nos anos de 2022 e 2023?

```python
# Contagem da ocorrência de cada tipo de incidente
tipo_counts = df_violencia_mulheres['CATEGORIA DOS INCIDENTES'].value_counts().reset_index()
tipo_counts.columns = ['CATEGORIA DOS INCIDENTES', 'contagem']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 8))

cor_principal = '#CDA2D7'  
cor_destaque = '#6D4C71'  
cor_fundo = '#EAEAEA'  
cor_texto = '#333333'  

palette = [cor_principal, cor_destaque]

sns.barplot(data=tipo_counts, x='contagem', y='CATEGORIA DOS INCIDENTES', palette=palette)

plt.title('Distribuição de Incidentes de Violência contra Mulheres por Código e Tipo em Vila Velha (2022-2023)', fontsize=16, color=cor_texto)
plt.xlabel('Número de Incidentes', fontsize=14, color=cor_texto)
plt.ylabel('Tipo de Incidente', fontsize=14, color=cor_texto)

plt.gca().set_facecolor(cor_fundo)

plt.tight_layout()  

plt.show()
```

### Qual é a distribuição de vítimas de violência contra mulheres em Vila Velha por faixa etária e tipo de incidente registrados em 2022 e 2023?

```python
# Agrupar e contar
faixa_etaria_counts = df_violencia_mulheres.groupby(['IDADE', 'CATEGORIA DOS INCIDENTES']).size().reset_index(name='contagem')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

faixa_etaria_counts['IDADE'] = pd.to_numeric(faixa_etaria_counts['IDADE'], errors='coerce')

faixa_etaria_counts = faixa_etaria_counts.dropna(subset=['IDADE'])

faixa_etaria_counts['Faixa Etária'] = pd.cut(faixa_etaria_counts['IDADE'], bins=range(0, 101, 10), right=False)

faixa_etaria_agrupada = faixa_etaria_counts.groupby(['Faixa Etária', 'CATEGORIA DOS INCIDENTES']).sum().reset_index()

plt.figure(figsize=(14, 8))

cor_principal = '#CDA2D7' 
cor_destaque = '#6D4C71'  
cor_fundo = '#EAEAEA'  
cor_texto = '#333333'  

palette = [cor_principal, cor_destaque]

sns.barplot(data=faixa_etaria_agrupada, x='Faixa Etária', y='contagem', hue='CATEGORIA DOS INCIDENTES', palette=palette)

plt.xticks(rotation=45, color=cor_texto)
plt.title('Distribuição de Vítimas de Violência contra Mulheres por Faixa Etária e Tipo de Incidente em Vila Velha (2022-2023)', fontsize=16, color=cor_texto)
plt.xlabel('Faixa Etária', fontsize=14, color=cor_texto)
plt.ylabel('Número de Vítimas', fontsize=14, color=cor_texto)

plt.gca().set_facecolor(cor_fundo)

plt.legend(title='Tipo de Incidente', title_fontsize='13', loc='center left', bbox_to_anchor=(1, 0.5), fontsize='11')

plt.tight_layout()  

plt.show()
```

### Como se distribuem as vítimas de violência contra mulheres em Vila Velha por raça e tipo de incidente, de acordo com os dados de 2022 e 2023?

### Qual é a quantidade de vítimas de violência contra mulheres em Vila Velha, distribuídas por gênero e tipo de incidente, nos anos de 2022 e 2023?

```python
# Criando uma tabela de frequência
tabela_frequencia_genero = df_violencia_mulheres.groupby(['SEXO', 'CATEGORIA DOS INCIDENTES']).size().reset_index(name='Contagem')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cores = ['#CDA2D7', '#6D4C71', '#EAEAEA', '#333333']  

tabela_frequencia_genero = df_violencia_mulheres.groupby(['SEXO', 'CATEGORIA DOS INCIDENTES']).size().reset_index(name='Contagem')

plt.figure(figsize=(12, 6))

paleta_personalizada = sns.color_palette(cores)

sns.barplot(data=tabela_frequencia_genero, x='SEXO', y='Contagem', hue='CATEGORIA DOS INCIDENTES', palette=paleta_personalizada)

plt.title('Quantidade de Vítimas de Violência contra Mulheres por Gênero e Tipo de Incidente em Vila Velha (2022-2023)', color='#333333', fontsize=16)
plt.xlabel('Gênero', color='#333333', fontsize=12)
plt.ylabel('Número de Vítimas', color='#333333', fontsize=12)

plt.xticks(rotation=45, color='#333333')
plt.yticks(color='#333333')

plt.legend(title='Categoria dos Incidentes', bbox_to_anchor=(1.05, 1), loc='upper left', title_fontsize=10, labelcolor='#333333')

plt.tight_layout()

plt.show()
```
