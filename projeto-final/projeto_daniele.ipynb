```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dados/base-youtube.csv', encoding='latin1')
df
```

```python
df = df.dropna(subset=['Country'])
df = df.dropna(subset=['category'])
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df_20_canais_inscritos = df[['Youtuber', 'subscribers', 'video views', 'category', 'Country']].sort_values(by='subscribers', ascending=False).head(20)
df_20_canais_inscritos
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_20_canais_inscritos['Youtuber'], df_20_canais_inscritos['subscribers'], marker='o', linestyle='-', color='g')
plt.xlabel('Canal Youtube')
plt.ylabel('Inscritos')
plt.title('Os 20 maiores Canais do Youtube')
plt.xticks(rotation=75)
plt.tight_layout()
plt.show()
```

```python
df_20_canais_visualizacoes = df[['Youtuber', 'subscribers', 'video views', 'category', 'Country']].sort_values(by='video views', ascending=False).head(20)
df_20_canais_visualizacoes
```

```python
df_canais_musica = df[df['category'] == 'Music']

df_rank_musica = df_canais_musica[['Youtuber', 'subscribers', 'video views', 'Country',]].sort_values(by='subscribers', ascending=False).head(10)

df_rank_musica.reset_index(drop=True, inplace=True)
df_rank_musica
```
